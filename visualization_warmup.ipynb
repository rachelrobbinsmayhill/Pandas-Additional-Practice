```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

url = 'https://gist.githubusercontent.com/zgulde/018f5d601efc79cb741a7aba92f51d29/raw/c08f1c09f605b43e9a354890f9dd68f296e76dc4/students.csv'
df = pd.read_csv(url)
```

2. `df` contains (fake) data on students time at codeup:

    - `ml_coffee`: coffee consumption in milliliters
    - `syntax_errors`: number of syntax errors
    - `p_days_absent`: percentage of days absent

```python
df.head()
```

3. Create a scatter plot of `syntax_errors` and `ml_coffee`. 
- Make sure to give your visual an appropriate title and labels.

```python
plt.scatter(df.syntax_errors, df.ml_coffee, alpha=0.5)

plt.xlabel('Syntax Errors', fontsize=15)
plt.ylabel('Coffee Consumed (ml)', fontsize=15)
plt.title('Syntax Errors in Relationship to Volume of Coffee Consumed', fontsize=20)

plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=df, x="syntax_errors", y="ml_coffee")

plt.xlabel('Syntax Errors', fontsize=15)
plt.ylabel('Coffee Consumed (ml)', fontsize=15)
plt.title('Syntax Errors in Relationship to Volume of Coffee Consumed', fontsize=20)

plt.tight_layout()
plt.show()
```

4. Create a scatter plot of `p_days_absent` and `ml_coffee`. Make sure to give your visual an appropriate title and labels.

```python
sns.scatterplot(data=df, x="p_days_absent", y= "ml_coffee")

plt.xlabel('Days Absent', fontsize=15)
plt.ylabel('Coffee Consumed (ml)', fontsize=15)
plt.title('Days Absent in Relationship to Volume of Coffee Consumed', fontsize=20)

plt.tight_layout()
plt.show()
```
