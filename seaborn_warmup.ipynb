# 2. Make a scatterplot of displacement vs horsepower. 

````python
ax = sns.scatterplot(x="displacement", y="horsepower", data=df)
````

Add the keyword argument `hue='weight'`.

````python
 ax = sns.scatterplot(x="displacement", y="horsepower", data=df, hue= "weight")
````

## What does this change? 

It shows the relationship between displacement and horsepower while controlling for or highlighting (delinitaing by) 'weight' with the different colors, so darker colors represent higher values.

## Add the keyword argument `palette='Greens'`. 

````python
 ax = sns.scatterplot(x="displacement", y="horsepower", data=df, hue= "weight", palette='Greens')
````

## What does this change?

This changes the color of the groupings from purples to greens.
